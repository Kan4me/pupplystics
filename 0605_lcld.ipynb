## Lethal Dose, Lethal Concentration

Acute toxicity describes the adverse effects of a substance that result either from a single exposure or from multiple exposures in a short period of time (usually less than 24 hours). To be described as acute toxicity, the adverse effects should occur within 14 days of the administration of the substance. (wiki)

Acute toxicity can be measured by:
- Lethal Dose LD50 for drug testing
- Lethal Concentration LC50 for environmental study

Here, we will focus on quality of the matadata based on the infos statistics and probability.
