# Previous analysis

The previous analysis led us to the understanding that species data may have **no** direct impact on relation of mortalily.
Based on that, it's interesting to narrow down the data base in order do have a more precice analysis.

````python
df = pd.read_csv("../data/treated_data/table.csv", dtype='unicode')
df = df.drop(['Unnamed: 0'], axis=1)
df.iloc[:5]
````

Changing the column name in order to do a query. ('class' is a python provate word and cannot be used in queries)

````python
df.rename(columns={'class':'animal_class'}, inplace=True)
df.iloc[:5]
````

Narrow down to Mammals with ```effect``` and ```measurement``` as mortality cathegory

````python
df_mammals = df.query('animal_class == "Mammalia"')
df_mammals = df.query('effect == "MOR"')
df_mammals = df.query('measurement == "MORT"')
df_mammals
````
